### Running The previous code on loop to get a live video stream

```python
import requests
import cv2
import numpy
```

```python
url="http://10.60.133.134:8080/shot.jpg"
```

```python
while True:
    data=requests.get(url)
    myphoto=data.content
    myphoto_a=bytearray(myphoto)
    photo_1d=numpy.array(myphoto_a)
    photo_3d=cv2.imdecode(photo_1d, -1)
    cv2.imshow('stream',photo_3d)
    if cv2.waitKey(1)==13:
        break
cv2.destroyAllWindows()
```
